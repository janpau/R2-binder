Riemannsum, rektangelmetoden, venstretiln√¶rming:

```python
a = 1
b = 5
n = 5000

def f(x):
    return x**2 - 2*x + 2

summen = 0
delta_x = (b-a)/n

for i in range(n):
    summen = summen + f(a + (i+1)*delta_x) * delta_x
    
print(round(summen,1))
```

Trapesmetoden:


```python
a = 1
b = 2
n = 10

def f(x):
    return 1/(x + x**2)

sum_m = 0
sum_v = 0
sum_h = 0
sum_t = 0

delta_x = (b-a)/n

for i in range(1,n+1): #midtpunkttiln√¶rming
    sum_m = sum_m + f(a + (i-1/2)*delta_x) * delta_x
    
for i in range(n): #venstretiln√¶rming
    sum_v = sum_v + f(a + i*delta_x) * delta_x
    
for i in range(n): #h√∏yretiln√¶rming
    sum_h = sum_h + f(a + (i+1)*delta_x) * delta_x
    
for i in range(n-1): #trapesmetoden
    sum_t = sum_t + f(a + (i+1)*delta_x)
    
sum_trapeser = (delta_x / 2) * (f(a) + f(b) + 2*sum_t)
       
print("Oppgave 2.28:")
print("Midtpunkttiln√¶rming:",round(sum_m,1))
print("Venstretiln√¶rming:",round(sum_v,1))
print("H√∏yretiln√¶rming:",round(sum_h,1))
print("Trapesmetoden:",round(sum_trapeser,1))
```
